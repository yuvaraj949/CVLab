# Step 3A: Image Transformations - For Jupyter Notebook

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
selfie = cv2.imread('my_selfie.jpg')
```

```python
# 1. Original
print("1. Original Image")
plt.figure(figsize=(5, 4))
plt.imshow(cv2.cvtColor(selfie, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original Image')
plt.show()
```

```python
# 2. Mirror (Horizontal Flip)
mirror = cv2.flip(selfie, 1)
print("2. Mirror Effect (Horizontal Flip)")
plt.figure(figsize=(5, 4))
plt.imshow(cv2.cvtColor(mirror, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Mirror Effect')
plt.show()
```

```python
# 3. Negative Effect
negative = 255 - selfie
print("3. Negative Effect")
plt.figure(figsize=(5, 4))
plt.imshow(cv2.cvtColor(negative, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Negative Effect')
plt.show()
```

```python
# 4. Brighten
bright = cv2.convertScaleAbs(selfie, alpha=1, beta=50)
print("4. Brightened Image (+50)")
plt.figure(figsize=(5, 4))
plt.imshow(cv2.cvtColor(bright, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Brightened +50')
plt.show()
```

```python
# 5. Darken
dark = cv2.convertScaleAbs(selfie, alpha=1, beta=-50)
print("5. Darkened Image (-50)")
plt.figure(figsize=(5, 4))
plt.imshow(cv2.cvtColor(dark, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Darkened -50')
plt.show()
```

```python
# 6. Rotate 45 degrees
h, w = selfie.shape[:2]
center = (w//2, h//2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_45 = cv2.warpAffine(selfie, M, (w, h))
print("6. Rotated 45")
plt.figure(figsize=(5, 4))
plt.imshow(cv2.cvtColor(rotated_45, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Rotated 45')
plt.show()
```

```python
# 7. Blur Effect
blurred = cv2.GaussianBlur(selfie, (15, 15), 0)
print("7. Blurred Image")
plt.figure(figsize=(5, 4))
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Blurred Image')
plt.show()
```

# Step 3B: Transformations Comparison Subplot

```python
# Step 2B: Transformations Comparison Subplot
import cv2
import numpy as np
import matplotlib.pyplot as plt

print(" Creating Transformations Comparison Subplot...")
print("=" * 50)

selfie = cv2.imread('my_selfie.jpg')

# Apply transformations
mirror = cv2.flip(selfie, 1)
negative = 255 - selfie
bright = cv2.convertScaleAbs(selfie, alpha=1, beta=50)
dark = cv2.convertScaleAbs(selfie, alpha=1, beta=-50)

# Rotation
h, w = selfie.shape[:2]
center = (w//2, h//2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_45 = cv2.warpAffine(selfie, M, (w, h))

# Blur
blurred = cv2.GaussianBlur(selfie, (15, 15), 0)

# Create transformations subplot
plt.figure(figsize=(15, 10))

# Original
plt.subplot(2, 4, 1)
plt.imshow(cv2.cvtColor(selfie, cv2.COLOR_BGR2RGB))
plt.title('1. Original')
plt.axis('off')

# Mirror
plt.subplot(2, 4, 2)
plt.imshow(cv2.cvtColor(mirror, cv2.COLOR_BGR2RGB))
plt.title('2. Mirror Flip')
plt.axis('off')

# Negative
plt.subplot(2, 4, 3)
plt.imshow(cv2.cvtColor(negative, cv2.COLOR_BGR2RGB))
plt.title('3. Negative')
plt.axis('off')

# Bright
plt.subplot(2, 4, 4)
plt.imshow(cv2.cvtColor(bright, cv2.COLOR_BGR2RGB))
plt.title('4. Brightened')
plt.axis('off')

# Dark
plt.subplot(2, 4, 5)
plt.imshow(cv2.cvtColor(dark, cv2.COLOR_BGR2RGB))
plt.title('5. Darkened')
plt.axis('off')

# Rotated
plt.subplot(2, 4, 6)
plt.imshow(cv2.cvtColor(rotated_45, cv2.COLOR_BGR2RGB))
plt.title('6. Rotated 45')
plt.axis('off')

# Blurred
plt.subplot(2, 4, 7)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.title('7. Blurred')
plt.axis('off')

plt.tight_layout()
plt.savefig('transformations_comparison.jpg', dpi=150, bbox_inches='tight')
plt.show()

print(" Saved transformations comparison: transformations_comparison.jpg")
```
