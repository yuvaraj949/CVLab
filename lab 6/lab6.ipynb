```python
# ðŸ§© Task 1 â€” Manual Computation of Normalized LoG Response
import numpy as np
import matplotlib.pyplot as plt

# Blob radius
r = 3

# Ïƒ values
sigma = np.arange(1, 6)

# Normalized LoG response
f = sigma**2 * np.exp(-(r**2) / (2 * sigma**2))

# --- Display table ---
print("Ïƒ  | Formula Substitution                | f(Ïƒ)")
print("---|-------------------------------------|-----------")
for s, val in zip(sigma, f):
    print(f"{s:<3}| {s}^2 * exp(-{r**2}/(2*{s}^2))      | {val:.6f}")

# --- Plot the response curve ---
plt.figure(figsize=(6,4))
plt.plot(sigma, f, 'o-', linewidth=2)
plt.title('LoG Response vs Scale (r = 3 px)')
plt.xlabel('Scale (Ïƒ)')
plt.ylabel('f(Ïƒ)')
plt.grid(True)
plt.show()

# --- Characteristic scale (where f(Ïƒ) is maximum) ---
max_sigma = sigma[np.argmax(f)]
print(f"\nMaximum response at Ïƒ = {max_sigma}")
```

```python
# ðŸ’» Task 2 â€” Verify Scale Invariance using Corrected Normalized LoG
import numpy as np
import matplotlib.pyplot as plt

# Blob radius
r = 3.0

# Continuous Ïƒ values for smoother curve
sigma = np.linspace(0.5, 6, 300)

# Corrected scale-normalized LoG response (proportional form)
g = (r**2 / sigma**2) * np.exp(-(r**2) / (2 * sigma**2))

# Find characteristic scale (maximum response)
max_sigma = sigma[np.argmax(g)]
max_val = np.max(g)

# --- Plot the response curve ---
plt.figure(figsize=(6,4))
plt.plot(sigma, g, 'b-', linewidth=2, label='Scale-normalized LoG')
plt.scatter([max_sigma], [max_val], color='r', label=f'Peak Ïƒ = {max_sigma:.2f}')
plt.title('Scale-normalized LoG Response vs Scale (r = 3 px)')
plt.xlabel('Scale (Ïƒ)')
plt.ylabel('Normalized Response')
plt.legend()
plt.grid(True)
plt.show()

print(f"Maximum normalized response at Ïƒ = {max_sigma:.4f}")
print(f"Theoretical characteristic scale (r / âˆš2) = {r/np.sqrt(2):.4f}")
```
