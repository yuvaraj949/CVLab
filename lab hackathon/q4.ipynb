```python
from IPython.display import Image, display
display(Image(filename='images/Q4a.jpg'))
display(Image(filename='images/Q4b.jpg'))

import numpy as np

# --- 1. Define Network Parameters ---
# Values extracted from Q4a.jpg and Q4b.jpg

# Input vector x (size 3)
# From Q4a.jpg: x = [1, 0.5, -1]
x = np.array([1.0, 0.5, -1.0])

# Hidden Layer Weights W1 (size 4x3) and Bias b1 (size 4)
# From Q4a.jpg
W1 = np.array([
    [0.5, 0.2, -0.1],
    [0.3, -0.5, 0.4],
    [-0.2, 0.1, 0.3],
    [0.4, -0.3, 0.2]
])

b1 = np.array([0.1, -0.2, 0.1, -0.1])

# Output Layer Weights W2 (size 3x4) and Bias b2 (size 3)
# From Q4b.jpg
W2 = np.array([
    [0.2, -0.1, 0.3, 0.1],
    [0.3, 0.2, -0.2, 0.4],
    [-0.1, 0.4, 0.2, -0.3]
])

b2 = np.array([0.1, -0.1, 0.2])

# --- 2. Define Helper Functions ---

def relu(z):
    return np.maximum(0, z)

def softmax(z):
    exp_z = np.exp(z - np.max(z)) # Numerical stability improvement
    return exp_z / np.sum(exp_z)

# --- 3. Forward Pass ---

# Layer 1 (Hidden)
z1 = np.dot(W1, x) + b1
h1 = relu(z1)

# Layer 2 (Output)
z2 = np.dot(W2, h1) + b2
probs = softmax(z2)

# Predicted Class
predicted_class_idx = np.argmax(probs)
predicted_class = f"Class {predicted_class_idx + 1}" # Assuming 1-based indexing

# --- 4. Print Results ---

print("Hidden Layer Activation (h1):\n", h1)
print("\nOutput Logits (z2):\n", z2)
print("\nSoftmax Probabilities:\n", np.round(probs, 4))
print("\nPredicted Class:", predicted_class)
```
